# Removendo moeda em feature

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Nome': ['Lucas', 'Pedro', 'Maria', 'Paulo', 'Eliana'],
                  'Salário': [3200.45, 'R$ 1550.00', 4200.80, 'R$1700.00', 4500.00]
                  })
```

```python
df
```

```python
df.info()
```

```python
df['Salário'] = df['Salário'].replace('[R$,]', '', regex=True).astype('float')
```
