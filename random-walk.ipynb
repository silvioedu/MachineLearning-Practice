## Random Walk

```
step_option = [1, -1]
step_choice = choice(step_option)
```

```
walk = []
walk.append(step_choice)
for step in range(1,1000):
  next_step = walk[step-1] + choice(step_option)
  walk.append(next_step)
print(walk)
```

```
plt.plot(walk)
plt.show()
```

## Ploting Random Walk but using functions

```
def random_walk(step_num):
  walk = []
  step_option = [1, -1]
  step_choice = choice(step_option)
  walk.append(step_choice)
  for step in range(1, step_num):
    walk.append(walk[step-1]+choice(step_option))
  return walk

def plot_walk(walk):
  x_axis = list(range(1, len(walk)+1))

  plt.plot(x_axis,walk)
  plt.title('Random walk')
  plt.xlabel('Number of steps')
  plt.ylabel('Distance')
  plt.show()
```

```
random_w = random_walk(100)
plot_walk(random_w)
```

## Creating few random walk in the same plot

```
def plot_walks(quantity, steps):

  lab_list = list(range(1,quantity+1))
  x = list(range(1, steps+1))
  plt.xlabel('Number of steps')
  plt.ylabel('Distance')
  plt.title(str(quantity) + ' random walks')

  for i in range(0,quantity):
    plot_label = 'Plot n. '+ str(lab_list[i])
    plt.plot(x, random_walk(steps), label = plot_label)
    plt.legend(loc='lower left')

  plt.show()
```

```
plot_walks(5,10000)
```
