# Visualizar maiores e menores

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    'Produto': ['Macarrão', 'Arroz', 'Feijão', 'Carne', 'Pão', 'Óleo', 'Farinha', 'Açúcar', 'Detergente', 'Sabão em Pó'],
    'Qtde': [12000, 8000, 10450, 7540, 10450, 6500, 3235, 800, 5670, 14518]
})
```

```python
df
```

```python
# Os maiores sem duplicidade da feature
df.nlargest(3, 'Qtde', keep='last')
```

```python
# Os maiores com duplicidade da feature
df.nlargest(3, 'Qtde', keep='all')
```

```python
# Os menores sem duplicidade da feature
df.nsmallest(3, 'Qtde', keep='last')
```

```python
# Os menores com duplicidade da feature
df.nsmallest(3, 'Qtde', keep='all')
```
