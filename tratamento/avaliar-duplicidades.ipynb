# Removendo duplicidades

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Produto': ['Macarrão', 'Arroz', 'Feijão', 'Carne', 'Pão', 'Óleo', 'Farinha', 'Arroz', 'Carne', 'Sabão em pó'],
    'Mogi-Mirim': [12000, 8000, 10450, 4589, 10450, 5400, 3235, 8000, 987, 14518],
    'Campinas': [784, 3459, 8453, 392, 934, 8500, 3835, 3459, 518, 1498],
    'São Paulo': [823, 6518, 870, 8918, 209, 7100, 7138, 6518, 163, 9183]
})
```

```python
df
```

```python
# Existem observações duplicadas?
df[df.duplicated(keep=False)]
```

```python
# Remover uma das linhas
df.drop_duplicates(ignore_index=True, inplace=True)
```

```python
df
```

```python
# Agrupando por uma feature
df[df.duplicated(keep=False, subset='Produto')]
```
