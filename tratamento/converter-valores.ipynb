```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Nome': ['Lucas', 'Pedro', 'Maria', 'Paulo', 'Eliana'],
    'Sexo': ['masculino', 'masculino', 'feminino', 'masculino', 'feminino'],
    'Fumante': ['não', 'não', 'sim', 'não', 'sim'],
    'Idade': [16, 19, 23, 17, 32]
})
```

```python
df
```

```python
df['Sexo MF'] = df['Sexo'].map({'masculino':'M', 'feminino': 'F'})
df['Fumante 01'] = df['Fumante'].factorize()[0]
df['Idade para dirigir'] = df['Idade'] >= 18
```
