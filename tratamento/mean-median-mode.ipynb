```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.DataFrame({
    'Anual income': [62000,64000,49000,324000,1264000,54330,64000,51000,55000,48000,53000]
})
```

```python
df
```

```python
# Cálculo das métricas
print("Média: {}".format(round(np.mean(df)[0])))
print("Mediana: {}".format(np.median(df)))
print("Mediana: {}".format(stats.mode(df)[0][0][0]))
```
