```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

%matplotlib inline
```

```python
df = pd.DataFrame({
    'Cidade': ['New York', 'Los Angeles', 'San Francisco'],
    'Vendas': [12327, 17129, 19923]
})
```

```python
df
```

```python
# Gr치fico de barras vertical
plt.bar("Cidade", "Vendas", data = df, color = "blue")
plt.xlabel("Cidades")
plt.ylabel("Ice cream")
plt.title("Sales")
plt.show()
```

```python
# Gr치fico de barras horizontal
plt.barh("Cidade", "Vendas", data = df, color = "red")
plt.xlabel("Cidades")
plt.ylabel("Ice cream")
plt.title("Sales")
plt.show()
```

```python
# Gr치fico de pizza
fig = plt.figure(figsize =(10, 7)) 
plt.pie(df['Vendas'], labels = df['Cidade']) 
plt.title('Sales')
plt.show() 
```

```python
# Gr치fico de Pareto
df = df.sort_values(by='Vendas',ascending=False)
df['% acumulada'] = round(df['Vendas'].cumsum()/df['Vendas'].sum()*100,2)

fig, ax = plt.subplots()
ax.bar(df.index, df['Vendas'], color="C0")
ax2 = ax.twinx()
ax2.plot(df['Cidade'], df['% acumulada'], color="C1", marker="D", ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())
ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
ax2.set_ylim([0,110])
plt.title('Sales')
plt.show()
df
```
