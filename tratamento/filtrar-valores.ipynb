# Filtrando valores

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Nome': ['Dayana', 'Camila', 'Felipe', 'Dayana', 'Jander', 'Paulo', 'Jander', 'Marcos', 'Oswaldo', 'Gabriela'],
                   'Cidade': ['Mogi-Mirim', 'Campinas', 'São Paulo', 'Campinas', 'Mogi-Mirim', 'Campinas', 'Goiânia', 'Itapira', 'Limeira', 'Mogi-Mirim']
                })
```

```python
# OR
df[(df['Nome'] == 'Jander') | (df['Nome'] == 'Dayana')]
```

```python
# AND
df[(df['Nome'] == 'Jander') & (df['Cidade'] == 'Mogi-Mirim')]
```

```python
# IN
df[df['Cidade'].isin(['Mogi-Mirin', 'Campinas'])]
```

```python
# NOT IN
df[~df['Cidade'].isin(['Mogi-Mirin', 'Campinas'])]
```

```python
# QUERY
df.query('Cidade == "Mogi-Mirim" or Nome == "Dayana"')
```
