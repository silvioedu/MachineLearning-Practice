```
from PIL import Image, ImageFilter, ImageFont, ImageDraw
from matplotlib.pyplot import imshow

print('Setup Complete')
```

```
# pre-visualizing image
file_name = '9.5.7_hillside_for_PIL_example.jpg'
im = Image.open(file_name)
imshow(im)
```

```
# image informations
print(f'size: {im.size} \nformat: {im.format_description} \nmode: {im.mode}')
```

```
# croping image
crop = (200, 200, 600, 600)
cropped_image = im.crop(crop)
imshow(cropped_image)
```

```
# image resized
resized_image = im.resize((400, 200))
imshow(resized_image)
```

```
# converting to gray scale
gray_scale = im.convert('LA')
imshow(gray_scale)
```

```
# adding gausshian blur
effect_image = im.filter(ImageFilter.GaussianBlur(radius=10))
imshow(effect_image)
```
